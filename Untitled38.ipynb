REPLACE THE MISSING VALUEs







```python
missingvalue=netflix.isna().sum()
missingvalue[missingvalue>0]
```

```python
netflix['director'].fillna('Unknown',inplace=True)
netflix['cast'].fillna('Unknown',inplace=True)
netflix['country'].fillna('Unknown',inplace=True)
```

```python
netflix['rating'].fillna(netflix['rating'].mode()[0],inplace=True)
```

```python
netflix['date_added']=pd.to_datetime(netflix['date_added'],errors='coerce')
```

```python
netflix['date_added'].fillna(netflix['release_year'],inplace=True)
```

```python
netflix.isna().sum() # clean dataset with no null values
```

 conversion of categorical attributes to 'category'

```python
netflix['type']=netflix['type'].astype('category')
netflix['rating']=netflix['rating'].astype('category')
netflix['country']=netflix['country'].astype('category')
```

UNNESTING : certain columns have multiple entries in one cell.So,we are unnesting columns director, cast, country, listed_in

```python
unnested_director=netflix.copy()
unnested_director['director']=unnested_director['director'].str.split(',')
unnested_director=unnested_director.explode('director')
unnested_director['director']=unnested_director['director'].str.strip()
unnested_director['director'].value_counts().head()
```

```python
unnested_cast=netflix.copy()
unnested_cast['cast']=unnested_cast['cast'].str.split(',')
unnested_cast=unnested_cast.explode('cast')
unnested_cast['cast']=unnested_cast['cast'].str.strip()
unnested_cast['cast'].value_counts().head()
```

* Anupam kher is the prominent actor featured in most of the content followed by Srk

```python
unnested_country = netflix.copy()
unnested_country['country'] = unnested_country['country'].str.split(",")
unnested_country = unnested_country.explode('country')
unnested_country['country'] = unnested_country['country'].str.strip()
unnested_country['country'].value_counts()
```

* United states followed by India contribute predominantly to the netflix viewership followed by UK

```python
unnested_listing = netflix.copy()
unnested_listing['listed_in'] = unnested_listing['listed_in'].str.split(",")
unnested_listing = unnested_listing.explode('listed_in')
unnested_listing['listed_in'] = unnested_listing['listed_in'].str.strip()
unnested_listing['listed_in'].value_counts().head()
```

```python
#creating seperate columns: movie duration and no of seasons for tv show

netflix.loc[netflix['type']=='Movie','movie_duration']=netflix.loc[netflix['type']=='Movie','duration'].str.extract('(\d+)')[0].astype('int')
netflix.loc[netflix['type']=='TV Show','no_of_seasons']=netflix.loc[netflix['type']=='TV Show','duration'].str.extract('(\d+)')[0].astype('int')
```

```python
netflix.isna().sum()
```

```python
netflix['movie_duration'].fillna(int(netflix['movie_duration'].mean()),inplace=True)
netflix['no_of_seasons'].fillna(int(netflix['no_of_seasons'].mean()),inplace=True)
```

```python
netflix.nunique()
```

Netflix has its own content rating system, in 6 tiers:

All (Young Children) (Television: TV-Y, TV-G) (Movies: G)

7+ (Children) (Television: TV-Y7) (Movies: soft PG)

10+ (Adolescents) (Television: TV-Y7-FV, soft TV-PG) (Movies: hard PG)

13+ (Teens) (Television: hard TV-PG, soft TV-14) (Movies: soft PG-13)

16+ (Older Teens) (Television: hard TV-14, TV-MA) (Movies: hard PG-13, R)

18+ (Adults) (Movies: NC-17, Unrated)

```python
netflix.groupby('rating')['movie_duration'].aggregate(['min','max','mean']).sort_values(by='mean',ascending=False)
```

* teen,adult and mature audience(NC-17,PG-13,R) receive longer duration movies while young children (TV-Y,TV-Y7-FV) receive shorter content

```python
netflix.groupby('rating')['no_of_seasons'].aggregate(['min','max','mean']).sort_values(by='mean',ascending=False)
```

* In Tv shows,younger audience receive more seasons comparatively to older and mature audience

```python
# How has the number of movies released per year changed over the last 20-30 years?
```

```python
movies=netflix[netflix['type']=='Movie']
movies_recent=movies[movies['release_year']>=1990]
movies_per_year=movies_recent.groupby('release_year')['title'].count()

sns.lineplot(x=movies_per_year.index,y=movies_per_year.values)
plt.title('Number of movies released per year')
plt.xlabel('Release year')
plt.ylabel('no of movies')
plt.show()
```

* A predominant growth in movie releases post 2000's and specifically since 2015 followed by significant decline later


```python
tvshows=netflix[netflix['type']=='TV Show']
tvshows_recent=tvshows[tvshows['release_year']>=1990]
tvshows_per_year=tvshows_recent.groupby('release_year')['title'].count()

sns.lineplot(x=tvshows_per_year.index,y=tvshows_per_year.values)
plt.title('Number of tvshows released per year')
plt.xlabel('Release year')
plt.ylabel('no of tvshows')
plt.show()
```

```python
# best time to launch a TV show
netflix['month_name']=netflix['date_added'].dt.month_name()
tv_show=netflix[netflix['type']=='TV Show']
plt.figure(figsize=(8,5))
sns.countplot(data=tv_show,x='month_name',order=tv_show['month_name'].value_counts().index)
plt.title('TV shows per month')
plt.xlabel('Month')
plt.ylabel('no of tv shows')
plt.xticks(rotation=90)
plt.show()
```

```python
#  Netflix has more focus on TV Shows or movies in recent years ?
netflix['year_added']=netflix['date_added'].dt.year
movies = netflix[netflix['type'] == 'Movie']
tv_shows = netflix[netflix['type'] == 'TV Show']
movies_per_year = movies.groupby('year_added').size().reset_index(name='count')
tv_shows_per_year = tv_shows.groupby('year_added').size().reset_index(name='count')

plt.figure(figsize=(5,5))
sns.lineplot(x='year_added', y='count', data=movies_per_year, label='Movies')
sns.lineplot(x='year_added', y='count', data=tv_shows_per_year, label='TV Shows')
plt.title('movies and tv shows added per year')
plt.show()
```

* In recent years,the focus was more on movie releases compared to tv shows
* probably since 2018, the content addition is seeing a decline

BUSINESS INSIGHTS:

*The primary content creation is focused significantly on movies and less on tv shows indicating a gap in tv show production and a potential area to tap

*The content is primarily focused on teen and adult audience which shows that target is lesser on family audience and kids

*The content creation has seen rapid increase post 2015 indicating focus on newer content in recent years

*In terms of countries,US is massively dominating in terms of content followed by India and Uk with almost no African representation in terms of content

*Mostly the movie duration is centered around ~ 100 minutes and and no of seasons per tv show is on avg 2 seasons and seasons increase for kids content

*A negative correlation between movie releases and duration shows that in recent years content is mostly of short duration type

*In terms of genres,international movies,tv shows along with crime and drama are doing well




RECOMMENDATIONS:

* Increase representation of content in emerging and growing markets like Africa,south korea and India

*Increase the content production of Tv shows as they are underrepresented comparatively to the movies

*Adding content centered to family and kids helps to cater to variety of viewers and not be marked as a streaming service meant for adults

*Need to invest and focus more on shorter duration content as movies with ~100 min duration dominate

*Adding country specific content based on popular genres will help to locate the target audience

*Certain actors and directors appear repeatedly in the content ...so leveraging and collaborating with them helps us provide the required content
